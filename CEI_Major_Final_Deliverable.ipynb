## Introduction:
Social equity means adjusting conditions to address people's diverse needs, strengths, and social realities so that everyone has the opportunity to thrive. Calgary Equity Index (CEI) measures equity across Calgary and is crucial in evaluating the equitable distribution of resources, opportunities, and services within Calgary. Calgary has 290 census tracts, 206 communities, 14 wards, and four quadrants (N.W., N.E., S.E., S.W.).
<br><br>
Each census tract is given ratings across 61 indicators grouped into 6 domains namely - Economic Opportunity, Human and Social Well being, Population Health, Community Belonging and Safety, Accessibility and Amenities, Climate and Environment. Lots of data is consolidated into a single index score. Higher total equity scores reflect areas of greater inequities and more overlapping barriers or needs. Lower total equity scores reflect areas with fewer inequities and fewer overlapping concerns. The CEI helps identify where gaps exist so that services, resources and support can be tailored to meet the specific needs of each census tract/community.
<br>

```python
# Load the CSV into a DataFrame
df = pd.read_csv("2024_CEI_Data.csv") 

# Create an in-memory SQLite database using SQLAlchemy
engine = create_engine('sqlite:///:memory:')

# Write the DataFrame to the SQLite database
df.to_sql('CEI_table', engine, index=False, if_exists='replace')  

# Run SQL queries 
query = "SELECT * FROM CEI_table"  
result = pd.read_sql_query(query, engine)

# Display the results
#print(result)
```

#### Query to get the Total_CEI Score and Domain scores for various communities

```python
query = """
SELECT CTUID, Communities, total_weighted_CEI_Score, climate_and_environment_domain_score, Quadrant, economic_opportunity_domain_score,
accessibility_and_amenities_domain_score, population_health_domain_score, community_belonging_and_safety_domain_score,
human_and_social_wellbeing_domain_score
FROM CEI_table

ORDER BY total_weighted_CEI_Score;
"""
Domainscores = pd.read_sql_query(query, engine)
```

## VISUALIZING THE INEQUITIES IN CALGARY'S QUADRANTS

We plot the CEI score vs the domain scores for all the communities.

```python
# Function to plot CEI score vs Domain scores showing Quadrants of the communities:

# columns_to_plot = {
#     'climate_and_environment_domain_score': 'Climate and Environment Domain Score',
#     'economic_opportunity_domain_score': 'Economic Opportunity Domain Score',
#     'accessibility_and_amenities_domain_score': 'Accessibility and Amenities Domain Score',
#     'population_health_domain_score': 'Population Health Domain Score',
#     'community_belonging_and_safety_domain_score': 'Community Belonging and Safety Domain Score',
#     'human_and_social_wellbeing_domain_score': 'Human and Social Wellbeing Domain Score'
# }

color_map = {
    'SW': 'blue',
    'NW': 'green',
    'SE': 'orange',
    'NE': 'red'
}

# Add a column to df to map the color
Domainscores['color'] = Domainscores['Quadrant'].map(color_map)


# function to plot CEI vs Domain Scores
def plot_CEI_Domain(column, domain_name):
    
    plt.figure(figsize=(10, 6))
    plt.scatter(
        Domainscores['total_weighted_CEI_Score'], 
        Domainscores[column], 
        c=Domainscores['color'],  # Set marker color based on quadrant
        alpha=0.7,
        label=df['Quadrant']
    )
    plt.title(f'CEI Score vs {domain_name}')
    plt.xlabel('CEI Score')
    plt.ylabel(domain_name)
    plt.grid(True)
    plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', label=quad, 
                markerfacecolor=color, markersize=10) for quad, color in color_map.items()])
    plt.show()
    

```

## HUMAN AND SOCIAL WELL BEING DOMAIN

The Human and Social Wellbeing domain score is an equally weighted combination of indicators that represent the diverse social and demographic characteristics of Calgarians. A high domain score reflects areas with greater inequities and more human and social wellbeing needs  A low domain score reflects areas with fewer inequities and fewer human and social wellbeing needs.

#### What are average domain scores for Human and Social Well Being Domain for the quadrants?

We observed that NE quadrant have a high score which implies that they have higher inequities in the domain of Human and Social Wellbeing. We investigated to find the common characteristics of these communities that are lagging behind.

```python
query = """
SELECT AVG(total_weighted_CEI_Score) AS Avg_CEI_score, AVG(human_and_social_wellbeing_domain_score) as Avg_Human_Social_Wellbeing_score, Quadrant
FROM CEI_table
GROUP BY Quadrant
ORDER BY Avg_CEI_score ASC;
"""
avg_climate_score = pd.read_sql_query(query, engine)
#result = pd.DataFrame(result)

print(avg_climate_score)
```

```python
query = """
SELECT  communities, quadrant, human_and_social_wellbeing_domain_score, Quadrant, 
        refugee_perc, recent_immigration_perc, indigenous_perc,
        seniors_65_perc, seniors_living_alone_perc, 
        single_parent_woman_perc,
        no_official_languages_perc, educ_no_cert_perc
FROM CEI_table
ORDER BY human_and_social_wellbeing_domain_score ASC
LIMIT 10;
"""
Least_Inequity_in_HS_Wellbeing = pd.read_sql_query(query, engine)
#print(Least_Inequity_in_HS_Wellbeing)
```

```python
# Query to calculate the average % of demographic sections from 10 communities with the least in-equities
query_new_table = """
SELECT AVG(refugee_perc) as Avg_Refugee_pct, 
       AVG(recent_immigration_perc) as Avg_Recent_Immi_pct, 
       AVG(indigenous_perc) as Avg_Indigenous_pct,
       AVG(single_parent_woman_perc) as Avg_Single_Parent_Woman_pct, 
       AVG(seniors_living_alone_perc) as Avg_Seniors_living_alone_pct,
       AVG(no_official_languages_perc) as Avg_No_offcial_Lang_pct,
       AVG(educ_no_cert_perc) as Avg_No_Education_pct
       
FROM  Least_Inequity_in_HS_Wellbeing;
""" 
avg_LowInequity_HS = pd.read_sql_query(query_new_table, engine)
#print(avg_LowInequity_HS)
```

```python
# Query to get the 10 communities with the most inequities in Human and Social Well being domain
query = """
SELECT  communities, human_and_social_wellbeing_domain_score, Quadrant, 
        refugee_perc, recent_immigration_perc, indigenous_perc,
        seniors_65_perc, seniors_living_alone_perc, 
        single_parent_woman_perc,
        no_official_languages_perc, educ_no_cert_perc
FROM CEI_table
ORDER BY human_and_social_wellbeing_domain_score DESC
LIMIT 10;
"""
Highest_Inequity_in_HS_Wellbeing = pd.read_sql_query(query, engine)

# Create a new table and store the results
Highest_Inequity_in_HS_Wellbeing.to_sql('Highest_Inequity_in_HS_Wellbeing', con=engine, if_exists='replace', index=False)
#print(Highest_Inequity_in_HS_Wellbeing)






#  Query to get the averages% of demographic sections from 10 communities with highest Ineqities in HS well being
query_new_table = """
SELECT AVG(refugee_perc) as Avg_Refugee_pct, 
       AVG(recent_immigration_perc) as Avg_Recent_Immi_pct, 
       AVG(indigenous_perc) as Avg_Indigenous_pct,
       AVG(single_parent_woman_perc) as Avg_Single_Parent_Woman_pct, 
       AVG(seniors_living_alone_perc) as Avg_Seniors_living_alone_pct,
       AVG(no_official_languages_perc) as Avg_No_offcial_Lang_pct,
       AVG(educ_no_cert_perc) as Avg_No_Education_pct
       
FROM  Highest_Inequity_in_HS_Wellbeing;
""" 
avg_HighInequity_HS = pd.read_sql_query(query_new_table, engine)



```

Climate and Environment Domain Score	The Climate and Environment domain score is an equally weighted combination of indicators that reflect climate risks and environmental conditions in Calgary. A high domain score reflects areas with greater inequities and more climate and environmental needs (such as areas with fewer trees, hotter temperatures, high flood risks, and more noise pollution). A low domain score reflects areas with fewer inequities and fewer climate and environmental needs (such as areas with greater tree density, lower flood risk, and more green spaces). ### What are the average Climate and Environment Domain Scores for Calgary's quadrants?

```python
# query to get the average Scores for Climate and Environment Domain:
query = """
SELECT AVG(total_weighted_CEI_Score) AS Avg_CEI_score, AVG(climate_and_environment_domain_score) as Avg_ClimateEnvironment_score, Quadrant
FROM CEI_table
GROUP BY Quadrant
ORDER BY Avg_CEI_score ASC;
"""
result = pd.read_sql_query(query, engine)
#result = pd.DataFrame(result)

print(result)
```
